```python
import json

# Specify the path to your JSON file
file_path = '/Users/reshmarajan/Desktop/RA/youtube videos/HANDS-ON VIDEO/11.Robot AI Demo.json'

# List to hold all texts
texts = []

# Open the file and read line by line
with open(file_path, 'r') as file:
    for line in file:
        # Parse the JSON data from each line
        json_data = json.loads(line.strip())
        
        # Extract the 'text' field and add it to the list
        texts.append(json_data['text'])

# Path to save the output text file
output_path = '/Users/reshmarajan/Desktop/Robot AI Demo_2.txt'

# Open the output text file and write each text to it, separated by one line space
with open(output_path, 'w') as output_file:
    for text in texts:
        output_file.write(text + '\n\n')  # Double newline for spacing

print("Texts have been saved to:", output_path)
```
