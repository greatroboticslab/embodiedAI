```python
from docx import Document
import re

# Load the source document
doc = Document("merged.docx")

# Extract YouTube links
youtube_links = []
for para in doc.paragraphs:
    matches = re.findall(r"https://www\.youtube\.com/watch\?v=[\w\-]+(?:[&?=+\w]*)*", para.text)
    youtube_links.extend(matches)

# Create a new document with just the links
new_doc = Document()
new_doc.add_heading("YouTube Links", level=1)
for link in youtube_links:
    new_doc.add_paragraph(link)

# Save the new document
new_doc.save("YouTube_Links.docx")
```
